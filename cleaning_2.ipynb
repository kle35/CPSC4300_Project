```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
data = pd.read_csv('airpod_reviews_cleaned.csv')
data.head()
```

```python
data['review_length'] = data['full_review'].apply(len)
data.head()
```

```python
%pip install vaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
def analyze_sentiment(text):
    analyzer = SentimentIntensityAnalyzer()
    sentiment = analyzer.polarity_scores(text)
    return sentiment
```

```python
data['sentiment'] = data['full_review'].apply(analyze_sentiment)
data.head()
```

```python
data['negative_sentiment_score'] = data['sentiment'].apply(lambda x: x['neg'])
data['neutral_sentiment_score'] = data['sentiment'].apply(lambda x: x['neu'])
data['positive_sentiment_score'] = data['sentiment'].apply(lambda x: x['pos'])
data['compound_sentiment_score'] = data['sentiment'].apply(lambda x: x['compound'])
data.head()
```

```python
data = data.drop(columns=['sentiment'])
```
